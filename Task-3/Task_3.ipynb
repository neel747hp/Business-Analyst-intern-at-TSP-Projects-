# Visualization of Dataset
## Profit Distibution Statewise

```python
plt.style.use('ggplot')
plt.figure(figsize=(30,15))
plt.bar('State','Profit', data=df, color="black")
plt.xticks(rotation=90)
plt.show()
```

# Here we found 
### Highest Profit State - Indiana
### Highest Loss State - Ohio

## Statewise customers count

```python
#print(df1['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.show()
```

Here, we found California has highest number of counts of customers

### Sub-category value count

```python
print(df['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'], color='red')
plt.xticks(rotation=90)
plt.show()
```

### Visulaization of number of segments and their counts

```python
print(df['Segment'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=df['Segment'])
plt.show()
```

```python
df.head()
```

### Visulaization of number of Categories and their counts

```python
sns.countplot(x=df['Category'])
plt.show()
```

```python
sns.pairplot(df, hue='Category', palette='hls')
```

## Heatmap of Data

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.lineplot('Category', 'Profit', data = df, color = 'green', label= 'Category')
plt.legend()
```

# Profit Vs sub-Categories

```python
plt.figure(figsize=(16,9))
plt.bar(df['Sub-Category'],df['Profit'], data=df, color="blue")
plt.xticks(rotation=90)
plt.show()
```
